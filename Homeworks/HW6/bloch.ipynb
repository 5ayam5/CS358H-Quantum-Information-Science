```python
import qutip as qt
import numpy as np
```

```python
bloch_sphere = qt.Bloch()
bloch_sphere.point_size = [200]
# bloch_sphere.point_marker = ['x']
bloch_sphere.point_color = ['black']
bloch_sphere.add_points([0, 0, np.cos(np.pi/4)])
bloch_sphere.add_vectors([[0, 0, 1], [0, 0, -1], [1, 0, 0], [-1/3, 0, 2*np.sqrt(2)/3]])
bloch_sphere.view = [210, 30]
bloch_sphere.show()
```
